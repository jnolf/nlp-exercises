```python
def parse_contact(person):
    output = {}
    output['name'] = person.find('h2').text
    output['quote'], output['email'], output['phone'], output['address'] = [p.text.strip() for p in person.find_all('p')]
    return output
```

```python
pd.DataFrame([parse_contact(person) for person in people])
```

```python
def get_codeup_blog (url):
    '''
    Function goes through Codeup blog links 
    '''
    # Set the headers
    headers = {'User-Agent': 'Codeup Data Science'}    
    # Get the http response object from the server
    response = get(url, headers=headers)
    # 
    soup = BeautifulSoup(response.text)   
    # Return a dictionary of all fields
    return{
    # What's the title of the blog
    'title': soup.select_one('.entry-title').text,
    # What are it's contents
    'content': soup.select_one('.entry-content').text.strip(),
    # When was it published
    'published': soup.select_one('.published').text
    } 
```

```python
def get_blog_articles (urls):
    # Create a list of dictionaries
    list_dic = [get_codeup_blog(url) for url in urls] 
    # Return a data frame
    return pd.DataFrame(list_dic).style.set_properties(**{'text-align':'left'})
```

```python
urls = ['https://codeup.com/workshops/from-bootcamp-to-bootcamp-a-military-appreciation-panel/',
        'https://codeup.com/featured/our-acquisition-of-the-rackspace-cloud-academy-one-year-later/',
        'https://codeup.com/workshops/virtual/learn-to-code-html-css-on-4-30/',
        'https://codeup.com/featured/5-books-every-woman-in-tech-should-read/',
        'https://codeup.com/cybersecurity/a-rise-in-cyber-attacks-means-opportunities-for-veterans-in-san-antonio/'    
       ]
```

```python
get_blog_articles(urls)
```
